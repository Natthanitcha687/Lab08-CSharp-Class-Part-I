## คำถาม

1. static class คืออะไร และแตกต่างจาก class ธรรมดาอย่างไร

*คำตอบ*

Static class ใน C# คือคลาสที่ไม่สามารถสร้าง instance ได้ และสมาชิกทั้งหมดของคลาสต้องเป็น static เช่นกัน ซึ่งแตกต่างจาก class ธรรมดาที่สามารถสร้าง instance ได้

Static class เหมาะสำหรับเก็บ utility methods หรือค่าคงที่ ที่ทำงานกับข้อมูลโดยรวมของโปรแกรม ในขณะที่ class ธรรมดาใช้สำหรับสร้าง object ที่มี state และ behavior เฉพาะตัว

2. เมื่อไหร่ควรใช้ static class และเมื่อไหร่ควรใช้ class ธรรมดา

*คำตอบ*

เลือกใช้ Static Class หากต้องการคลาสที่ทำงานกับข้อมูลส่วนกลางของโปรแกรม โดยไม่ต้องสร้างอ็อบเจ็กต์

เลือกใช้ Class ธรรมดา หากต้องการสร้างอ็อบเจ็กต์ที่มีสถานะและพฤติกรรมที่แตกต่างกัน

3. สมาชิกทั้งหมดของ static class ต้องเป็น static หรือไม่ เพราะอะไร

*คำตอบ*

ใช่ สมาชิกทั้งหมดของ static class ต้องเป็น static เนื่องจาก static class ไม่สามารถสร้าง instance ได้ ดังนั้นสมาชิกใดๆ ที่ไม่ใช่ static จะไม่สามารถเข้าถึงได้

4. เราสามารถสร้าง instance ของ static class ได้หรือไม่  เพราะอะไร

*คำตอบ*

ไม่ได้ เนื่องจาก static class ใน C# ไม่สามารถสร้าง instance ได้ เพราะ static class ถูกออกแบบมาให้มีสมาชิกทั้งหมดเป็น static ซึ่งหมายความว่าสมาชิกเหล่านั้นเป็นของคลาสโดยตรง ไม่ได้ผูกกับ instance ใดๆ
