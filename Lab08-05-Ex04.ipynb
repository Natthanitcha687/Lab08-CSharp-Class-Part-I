## คำถาม

1. Field ในภาษา C# คืออะไร


*คำตอบ*

ตัวแปรที่ถูกประกาศภายใน class หรือ struct และใช้สำหรับเก็บค่าของวัตถุ

2. Field มีหน้าที่อะไรใน class 


*คำตอบ*

ใช้เก็บข้อมูลของ instance (วัตถุ) หรือ class และสามารถกำหนดการเข้าถึง (access modifier) ได้

3. การประกาศ Field ทำได้อย่างไร

*คำตอบ*

การประกาศ Field ทำได้โดยการระบุชนิดข้อมูล ตามด้วยชื่อ Field และอาจกำหนดค่าเริ่มต้นหรือไม่ก็ได้

4. Implicit fields initialization  คืออะไร

*คำตอบ*

Implicit fields initialization คือการที่ Field ได้รับค่าเริ่มต้นเป็นค่าเริ่มต้นของชนิดข้อมูลนั้นๆ โดยอัตโนมัติ หากไม่ได้กำหนดค่าเริ่มต้นให้ Field ตอนที่ประกาศ

5. Explicit field initialization คืออะไร



*คำตอบ*

Explicit field initialization คือการที่ Field ได้รับการกำหนดค่าเริ่มต้นโดยตรงตอนที่ประกาศ โดยใช้เครื่องหมายเท่ากับ (=) ตามด้วยค่าที่ต้องการ

6. Implicit fields initialization  และ Explicit field initialization แตกต่างกันอย่างไร



*คำตอบ*

Implicit field initialization เหมาะสำหรับ Field ที่ต้องการค่าเริ่มต้นเป็นค่าเริ่มต้นของชนิดข้อมูล ในขณะที่ Explicit field initialization เหมาะสำหรับ Field ที่ต้องการค่าเริ่มต้นที่แตกต่างจากค่าเริ่มต้นของชนิดข้อมูล
