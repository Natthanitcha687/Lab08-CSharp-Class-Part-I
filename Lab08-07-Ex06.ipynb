## คำถาม

1. Method with parameter คืออะไร


*คำตอบ*

Method with parameter คือ เมธอดที่รับค่า input หรือข้อมูลจากภายนอก เพื่อนำไปประมวลผลหรือใช้งานภายในเมธอด ทำให้เมธอดมีความยืดหยุ่นและสามารถทำงานได้หลากหลายรูปแบบมากขึ้นอยู่กับค่าที่รับเข้ามา

2. ทำไมเราต้องใช้ parameter ใน method


*คำตอบ*

ช่วยให้เมธอดทำงานได้หลากหลายรูปแบบ โดยรับค่าที่แตกต่างกัน สามารถสร้างเมธอดเดียวที่ทำงานได้หลากหลายรูปแบบ แทนที่จะต้องสร้างหลายเมธอด ช่วยให้โค้ดอ่านง่าย เข้าใจง่าย และบำรุงรักษาง่ายขึ้น

3. parameter และ argument แตกต่างกันอย่างไร


*คำตอบ*

Parameter: ตัวแปรที่ประกาศไว้ในส่วนหัวของเมธอด เพื่อรับค่าจากภายนอก
Argument: ค่าที่ส่งเข้าไปในเมธอดเมื่อเรียกใช้งาน

4. ส่วนประกอบของ method ที่มี parameter มีอะไรบ้าง


*คำตอบ*

8. จงเขียน method ที่ตรวจสอบว่าตัวเลขที่รับเข้ามาเป็นจำนวนคู่หรือจำนวนคี่ และแสดงผลลัพธ์ทางหน้าจอ

````
public static void CheckEvenOdd(int number)
{
    if (number % 2 == 0)
    {
        Console.WriteLine(number + " เป็นจำนวนคู่");
    }
    else
    {
        Console.WriteLine(number + " เป็นจำนวนคี่");
    }
}

// ตัวอย่างการเรียกใช้เมธอด
CheckEvenOdd(10); // Output: 10 เป็นจำนวนคู่
CheckEvenOdd(7);  // Output: 7 เป็นจำนวนคี่
````
