## คำถาม

1. Access modifier ในภาษา C# คืออะไร


*คำตอบ*

ตัวกำหนดสิทธิ์ ในการเข้าถึงส่วนต่างๆ ของคลาส เช่น field, method, property ว่าอนุญาตให้ใครเข้าถึงได้บ้าง

2. จงอธิบายขอบเขตและการใช้งานของ Access modifier ต่อไปนี้ 


2.1  public


*คำตอบ*

ขอบเขต: เข้าถึงได้จากทุกที่

การใช้งาน: ใช้กับสมาชิกที่ต้องการให้ทุกคนเข้าถึงได้ เช่น method ที่เป็น interface ของคลาส

2.2 private


*คำตอบ*

ขอบเขต: เข้าถึงได้เฉพาะภายในคลาสตัวเองเท่านั้น

การใช้งาน: ใช้กับสมาชิกที่ไม่ต้องการให้ภายนอกคลาสเข้าถึงได้ เช่น field ที่เก็บข้อมูลภายใน

2.3 protected


*คำตอบ*

ขอบเขต: เข้าถึงได้ภายในคลาสตัวเองและคลาสที่สืบทอด (derived class)

การใช้งาน: ใช้กับสมาชิกที่ต้องการให้คลาสลูกเข้าถึงได้ แต่ไม่ต้องการให้ภายนอกเข้าถึงโดยตรง

2.4 internal


*คำตอบ*

ขอบเขต: เข้าถึงได้เฉพาะภายใน assembly เดียวกัน (โปรเจกต์เดียวกัน)

การใช้งาน: ใช้กับสมาชิกที่ต้องการให้ใช้ได้เฉพาะภายในโปรเจกต์เดียวกันเท่านั้น

2.5 protected internal

*คำตอบ*

ขอบเขต: เข้าถึงได้ภายใน assembly เดียวกัน หรือคลาสที่สืบทอดมาจากคลาสนี้ แม้จะอยู่คนละ assembly

การใช้งาน: ใช้เมื่อต้องการให้คลาสลูกที่อยู่ต่างโปรเจกต์กันเข้าถึงได้
