## คำถาม

1. Field ในภาษา C# คืออะไร


*คำตอบ*

ตัวแปรที่ถูกประกาศภายใน class หรือ struct และใช้สำหรับเก็บค่าของวัตถุ

2. Field มีหน้าที่อะไรใน class 


*คำตอบ*

ใช้เก็บข้อมูลของ instance (วัตถุ) หรือ class และสามารถกำหนดการเข้าถึง (access modifier) ได้

3. การประกาศ Field ทำได้อย่างไร ให้อธิบายรูปแบบการประกาศ field ในคลาส


*คำตอบ*

รูปแบบทั่วไปของการประกาศ Field ในคลาส:

class ExampleClass  
{  
    public int publicField;  // Field แบบ public  
    private string privateField; // Field แบบ private  
 
    public static int staticField;  
}  

4. Field และ Variable แตกต่างกันอย่างไร


*คำตอบ*

Field คือตัวแปรที่ประกาศไว้ในคลาสหรือโครงสร้าง (struct) เพื่อเก็บข้อมูลที่เกี่ยวข้องกับออบเจ็กต์หรืออินสแตนซ์ของคลาสนั้นๆ Field เป็นส่วนหนึ่งของออบเจ็กต์และมีอายุการใช้งานเท่ากับออบเจ็กต์นั้นๆ

Variable คือตัวแปรที่ประกาศไว้ในเมธอดหรือบล็อกของโค้ด เพื่อเก็บข้อมูลที่ใช้เฉพาะภายในขอบเขต (scope) ของเมธอดหรือบล็อกนั้นๆ Variable มีอายุการใช้งานจำกัดเฉพาะภายในขอบเขตที่ประกาศ

5. เมื่อไหร่ควรใช้ Field และเมื่อไหร่ควรใช้ Variable

*คำตอบ*

ใช้ Field เมื่อต้องการเก็บข้อมูลที่เกี่ยวข้องกับออบเจ็กต์และต้องการให้ข้อมูลนั้นอยู่กับออบเจ็กต์ไปตลอด เช่น ชื่อ อายุ หรือข้อมูลอื่นๆ ของบุคคล
ใช้ Variable เมื่อต้องการเก็บข้อมูลที่ใช้เฉพาะภายในเมธอดหรือบล็อกของโค้ด เช่น ตัวแปรที่ใช้ในการคำนวณ หรือตัวแปรที่ใช้ในการวนลูป
